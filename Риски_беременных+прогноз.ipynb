```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.metrics import classification_report
```

```python
data = pd.read_csv('Maternal Health Risk Data Set.csv')
data.head()
```

```python
X = data.drop(['RiskLevel'], axis=1)
y = data['RiskLevel']
```

```python
model = RandomForestClassifier()
model.fit(X,y)
```

```python
new_user_data = pd.DataFrame({
    'Age': [25],                  # Возраст в годах
    'SystolicBP': [120],          # Верхнее значение артериального давления в мм рт.ст.
    'DiastolicBP': [80],          # Нижнее значение артериального давления в мм рт.ст.
    'BS': [5.5],                  # Уровень глюкозы в крови в ммоль/л
    'BodyTemp' : [98.0],           # предположим, что температура тела пользователя 98.6 градусов по Фаренгейту
    'HeartRate': [70],            # Частота сердечных сокращений в ударах в минуту
})
predicted_risk = model.predict(new_user_data)
print("Предсказанный риск беременности:", predicted_risk)
#predicted_risk
```
