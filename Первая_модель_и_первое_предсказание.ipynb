```
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
```

```

df = pd.read_excel('trips_data_for_ML.xlsx', index_col = 0)
df.head()
```

```
X = df.drop('target', axis = 1)
y = df['target']
```

```
X.head()
```

```
y.head(8)
```

```
y.value_counts()
```

```

model = RandomForestClassifier()
```

```
model.fit(X, y)  # обучаем модель
```

```
# меняйте данные в перемернной example и делайте предсказания
example = {
    'salary': [90000],
    'age': [20],
    'family_members': [2],
    'city_Екатеринбург': [0],
    'city_Киев': [0],
    'city_Краснодар': [0],
    'city_Минск': [1],
    'city_Москва': [0],
    'city_Новосибирск': [0],
    'city_Омск': [0],
    'city_Петербург': [0],
    'city_Томск': [0],
    'city_Хабаровск': [0],
    'city_Ярославль': [0],
    'vacation_preference_Архитектура': [0],
    'vacation_preference_Ночные клубы': [0],
    'vacation_preference_Пляжный отдых': [1],
    'vacation_preference_Шоппинг': [1],
    'transport_preference_Автомобиль': [1],
    'transport_preference_Космический корабль': [0],
    'transport_preference_Морской транспорт': [0],
    'transport_preference_Поезд': [0],
    'transport_preference_Самолет': [0]}
example_df = pd.DataFrame(example)
predicted_proba = model.predict_proba(example_df)
predicted_class_index = np.argmax(predicted_proba, axis=1)
predicted_class = model.classes_[predicted_class_index]
print(predicted_class)
```
